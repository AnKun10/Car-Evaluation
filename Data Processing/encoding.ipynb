```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import matplotlib as plt
from category_encoders import TargetEncoder, OneHotEncoder
import seaborn as sns

df = pd.read_csv('Dao_version_price_final.csv')
print(df.head())
```

```python
count = df['grade'].nunique()
print(count)
```

```python
# Khởi tạo encoder với cột 'grade'
encoder = TargetEncoder(cols=['grade'])

# Fit encoder với dữ liệu
encoder.fit(df['grade'], df['price_in_billion'])

# Áp dụng encoder cho cột 'grade'
df['grade'] = encoder.transform(df['grade'])
print(len(df['grade'].unique()))
```

```python
columns_to_encode = ['grade']
one_hot_matrix = pd.get_dummies(df['grade'])

# Chuyển đổi các giá trị 'True' và 'False' thành '1' và '0'
one_hot_matrix = one_hot_matrix.astype(int)
print(one_hot_matrix)
```

```python
one_hot_matrix = pd.get_dummies(df['grade'])

# Gộp ma trận one-hot vào DataFrame
df = pd.concat([df, one_hot_matrix], axis=1)
```
