```python
import pandas as pd
```

```python
df = pd.read_csv('new_car_detail_en.csv')
df.head(1)
df = df.loc[:, ['ad_id', 'brand', 'grade', 'car_name']]
df.head()
```

```python
# Processing for grade name:
# Aura

current_brand = df[df['brand'] == 'Acura']

# Trích từ cột 'car_name' ra từ kí tự đầu tiên cho đến ngay trước kí tự số đầu tiên
car_names_trimmed = current_brand['car_name'].str.extract(r'(^.*?)(?=\d)')
car_names_trimmed = car_names_trimmed.replace(r'^\S+\s*', '', regex=True)
# Thêm cột mới vào DataFrame và gán giá trị từ phần được trích xuất
df = df.assign(grade=car_names_trimmed)
current_brand = current_brand.assign(grade = car_names_trimmed)

# In ra DataFrame sau khi thêm cột mới
current_brand.head()
```

```python
# Processing for grade name:
# Bentley

current_brand = df[df['brand'] == 'Bentley']

# Trích từ cột 'car_name' ra từ kí tự đầu tiên cho đến ngay trước kí tự số đầu tiên
car_names_trimmed = current_brand['car_name'].str.extract(r'(^.*?)(?=\d)')
car_names_trimmed = car_names_trimmed.replace(r'^\S+\s*', '', regex=True)
car_names_trimmed = car_names_trimmed.replace('V', '', regex= True)
# Thêm cột mới vào DataFrame và gán giá trị từ phần được trích xuất
df = df.assign(grade=car_names_trimmed)
current_brand = current_brand.assign(grade = car_names_trimmed)

# In ra DataFrame sau khi thêm cột mới
current_brand
```
