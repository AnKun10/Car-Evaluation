```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import matplotlib as plt
from category_encoders import TargetEncoder, OneHotEncoder
import seaborn as sns

df = pd.read_csv('Dao_version_price_final.csv')
df.head(1)
```

```python
df = df.drop('brand',axis=1)
```

```python
df = df.drop(["ad_id",], axis=1)
df.head()
```

```python
# processing 'origin','condition' column
df['origin'].replace({'Domestic assembly': 1, 'Imported': 0}, inplace=True)
df['condition'].replace({'New car': 1, 'Used car':0}, inplace=True)
df['transmission'].replace({'Manual':1, 'Automatic':0})
df.head()
```

```python
# use one-hot encoder for other columns
encoder = OneHotEncoder(cols=['car_model', 'exterior_color', 'interior_color', 'drive_type', 'engine','grade','transmission'])
encoded_df = encoder.fit_transform(df)
```

```python
encoded_df.head()
```

```python
# check data types of encoded_df
print(encoded_df.dtypes)
```

```python
# check if any columns is not numerical
non_numeric_columns = encoded_df.select_dtypes(exclude=['int', 'float']).columns
print(non_numeric_columns)
```

```python
null_rows = encoded_df[encoded_df.isnull().any(axis=1)]

print("Các hàng chứa giá trị null:")
print(null_rows)
```
